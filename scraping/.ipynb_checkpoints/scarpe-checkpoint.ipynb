```python
import json
```

```python
with open('links.json', 'r') as f:
    data = json.load(f)
```

```python
import re
```

```python
pattern = re.compile(r'\d{5,}')
```

```python
id = []
for i in data:
    for link in i:
        match = pattern.search(link)
        if match:
            id.append(int(match.group()))
```
