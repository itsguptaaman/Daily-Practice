```python
# Calculate total weight and applicable weight slab
merged_data['Total Weight'] = merged_data['Order Qty'] * merged_data['Weight (g)'] / 1000  # Convert grams to kilograms
```

```python
# Define a function to calculate the applicable weight slab
def calculate_applicable_weight_slab(row):
    zone = row['Zone']
    total_weight = row['Total Weight']
    
    # Retrieve applicable weight slab from the rate card data
    rate_card_zone = courier_rate_card_data[courier_rate_card_data['Zone'] == zone]
    
    if not rate_card_zone.empty:
        applicable_slab = rate_card_zone.iloc[0]['Weight Slab']
        return applicable_slab
    
    # Return a default value if no matching rate card information is found
    return 0  # Replace with an appropriate default value
```

```python
# Apply the function to calculate the applicable weight slab for each row
merged_data['Applicable Weight Slab'] = merged_data.apply(calculate_applicable_weight_slab, axis=1)
```

```python
forward_fixed = courier_rate_card_data['Forward Fixed Charge']
forward_additional = courier_rate_card_data['Forward Additional Weight Slab Charge']
rto_fixed = courier_rate_card_data['RTO Fixed Charge']
rto_additional = courier_rate_card_data['RTO Additional Weight Slab Charge']
total_weight = courier_invoice_data['Charged Weight']
applicable_slab = merged_data['Applicable Weight Slab']
```

```python
# Calculate the expected charges based on the rate card
def calculate_expected_charges(row):    
    expected_charges = forward_fixed
    if total_weight > applicable_slab:
        expected_charges += (total_weight - applicable_slab) * forward_additional
    expected_charges += rto_fixed
    if total_weight > applicable_slab:
        expected_charges += (total_weight - applicable_slab) * rto_additional
    
    return expected_charges
```

```python
# Apply the function to calculate the expected charges for each row
merged_data['Expected Charges'] = merged_data.apply(calculate_expected_charges, axis=1)

# Calculate the difference between expected charges and billed charges
merged_data['Difference'] = merged_data['Expected Charges'] - merged_data['Billing Amount (Rs.)']

# Export the calculated data to an Excel file
merged_data.to_excel('output.xlsx', index=False)
```
